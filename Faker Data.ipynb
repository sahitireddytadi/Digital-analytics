```python
import csv
from faker import Faker
import datetime

def datagenerate(records, headers):
    fake = Faker('en_US')
    fake1 = Faker('en_GB')   # To generate phone numbers
    with open("order_fk.csv", 'wt') as csvFile:
        writer = csv.DictWriter(csvFile, fieldnames=headers)
        writer.writeheader()
        orderid = [f for f in range(1007,1900)] 
        for j,i in enumerate(range(records)):
            totalitem=fake.random_int(1,10)
            shippingfee = fake.random_int(5,30)
            tax=round(fake.random.random(),2)
            totalcost=round(fake.random.uniform(20,100),2)
            orderdate=fake.date_between(start_date=datetime.date(2018,10,1), end_date=datetime.date(2018,10,31))
            deliverydate=fake.date_between(start_date=datetime.date(2018,11,1), end_date=datetime.date(2018,12,31))
            trackingnumber=fake.bban()
            deliverystatus=fake.random_int(0,1)
            
           
            writer.writerow({
                    "order_id" : orderid[j],
                    "total_item" : totalitem,
                    "shipping_fee": shippingfee,
                    "tax": tax,
                    "total_cost" : totalcost,
                     "order_date": orderdate,
                "delivery_date":deliverydate,
                 "tracking_number": trackingnumber,
                    "delivery_status":deliverystatus,
           
                    })
    
if __name__ == '__main__':
    records = 893
    headers = ["order_id", "total_item", "shipping_fee", "tax", "total_cost","order_date","delivery_date","tracking_number","delivery_status",]
    datagenerate(records, headers)
    print("CSV generation complete!")
```
