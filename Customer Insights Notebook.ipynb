```python
#identify and separate big spenders, lots of orders for
#sales and marketing campaign use; need to be separate flags because they aren't all mutually
#exclusive

def big_spender(row):
    if row['total_cost'] >= 5026.05:
        return 'Yes'
    else:
        return 'No'

def many_orders(row):
    if row['total_item'] >= 10:
        return 'Yes'
    else:
        return 'No'

user_data['big_spender'] = user_data.apply(big_spender, axis=1)
user_data['many_orders'] = user_data.apply(many_orders, axis=1)
```
